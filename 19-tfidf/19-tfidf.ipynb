## качаем данные

```python
# список ссылок на тексты
links = {
    'Rolling Stones - Paint it BlackYouTube': 'https://music.yandex.ru/translate/track/1703893',
    'Creedence - Suzie Q.': 'https://music.yandex.ru/translate/track/37408169',
    'The Doors - Riders on the Storm': 'https://music.yandex.ru/translate/track/985760',
    'Black Sabbath - Paranoid': 'https://music.yandex.ru/translate/track/645247',
'The Doors - People Are Strange': 'https://music.yandex.ru/translate/track/17190963',
}
```

```python
class HttpGetRequest:

    def __init__( self, header_params=None, ):
        self._header_params = header_params
        
    def get(self, url, params=None):
        try:
            response = requests.get(url,headers=self._header_params,params=params,)
            response.raise_for_status() 
            logging.debug('HttpGetRequest: request success')
            return response

        except HTTPError as http_err:
            logging.error(f'HttpGetRequest: request : {http_err}')

        except Exception as err:
            logging.error(f'HttpGetRequest: request : {err}')

        return ''    


class YandexSongDownloader(HttpGetRequest):
    
    def __init__( self ):
        # header_params = {  'x-auth-token': token, 'Content-Type': 'application/x-www-form-urlencoded', }  
        header_params = {
             'Content-Type':'text/html; charset=utf-8', 
             'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; rv:91.0) Gecko/20100101 Firefox/91.0',
             }
        super().__init__(header_params=header_params)
        
        
    def get(self,url):
            
        return  (
            BeautifulSoup( super().get(url).text, "html.parser")
            .find('div', class_='container song-text')
            .find('pre')
            .text
        ) 
```

```python
# качаем тексты
data = pd.DataFrame({
        t: [  YandexSongDownloader().get(links[t]), ]
        for t in links
    }).T
data.columns = ['text']

display(data)
```

```python
assert (data['text'].str.len()>1).all()
```

## частотный анализ

```python
# токенайзер - режем текст на отдельные слова

from nltk.tokenize import word_tokenize as nltk_tokenize_word
from nltk.corpus import stopwords as nltk_stopwords
from nltk.stem import WordNetLemmatizer
import re

stopwords = set(nltk_stopwords.words('russian'))
lemmatizer = WordNetLemmatizer()

def tokenizer(text,stopwords=stopwords,lemmatizer=lemmatizer):
    return [ 
        lemmatizer.lemmatize(t)
        for t in nltk_tokenize_word( text.lower(), language='english' ) 
        if (t not in stopwords) and ( re.match(r'^[A-z]{3,}$',t) ) 
    ] 
```

```python
# рисуем таблицу результата
def tf_table(x,v,d):
    print( 'словарь:', len(v), 'слов\n') 
    return pd.concat([ 
        pd.DataFrame(x,columns=v), 
        pd.Series(d,name='Song')
    ],axis=1).set_index('Song')
```

---

```python
# простой частотный анализ - считаем слова

from sklearn.feature_extraction.text import CountVectorizer

tf_model = CountVectorizer(
        tokenizer=tokenizer, # ф-ция токенайзер
        token_pattern=None, # отключаем дефолтный токенайзер
    ).fit(data['text'])

tf_table(
        x = tf_model.transform(data['text']).todense(),
        v = sorted( tf_model.vocabulary_.keys() ),
        d=data.index
    )  # .style.background_gradient(cmap='plasma', ) 
```

----

```python
# частотный анализ с нормализацией и обратной частотой (TF-IDF)

from sklearn.feature_extraction.text import TfidfVectorizer

# использования токенайзера вместе с векторайзером
tf_model = TfidfVectorizer(
        min_df=.01, # выкидываем очень редкие слова
        max_df=.50, # выкидываем очень частые слова
        use_idf=True, # используем обратную частоту
        norm='l2', # нормируем TF
        tokenizer=tokenizer, # ф-ция токенайзер
        token_pattern=None, # отключаем дефолтный токенайзер
    ).fit( data['text'] )


tf_table(
    x = tf_model.transform(data['text']).todense(),
    v = sorted( tf_model.vocabulary_.keys() ),
    d=data.index
)
```
